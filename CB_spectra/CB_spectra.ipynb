```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
font = {'family' : 'sans-serif',
            'size'   : 35}
plt.rc('font', **font)
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
```

# $\hat{C}_L^{\alpha\alpha}$ from PR3 data products$\hat{C}_L^{\alpha\alpha}$ from PR3 data products

```python
#file.csv containing multipoles ranging from L=0 up to L=1500 and the four CB spectra presented in Zagatti et al.
PR3_data = pd.read_csv('PR3_CB_spectra.csv')

#available spectra
CB_spectra = ['Full mission', 'HM1 x HM1', 'HM2 x HM2', 'HM1 x HM2']
```

```python
#visualization example of the CB power spectrum from Planck PR3 full mission data products
plt.figure(figsize=(15,11))
plt.plot(PR3_data['L'], PR3_data[CB_spectra[0]], color='black', label=CB_spectra[0])
plt.xlim(0,1500)
plt.ylim(-0.02, 0.02)
plt.xlabel('L')
plt.ylabel(r'$\hat{\mathcal{C}}_L^{\alpha\alpha}\;[\rm{deg}^2]$')
plt.legend()
plt.grid()
```

# $\hat{C}_L^{\alpha\alpha}$ from NPIPE data products

```python
#file.csv containing multipoles ranging from L=0 up to L=1500 and the four CB spectra presented in Zagatti et al.
NPIPE_data = pd.read_csv('NPIPE_CB_spectra.csv')

#available spectra
CB_spectra = ['Full mission', 'A x A', 'B x B', 'A x B']
```

```python
#visualization example of the CB power spectrum from Planck NPIPE splitA data products
plt.figure(figsize=(15,11))
plt.plot(NPIPE_data['L'], NPIPE_data[CB_spectra[1]], color='black', label=CB_spectra[1])
plt.xlim(0,1500)
plt.ylim(-0.04, 0.04)
plt.xlabel('L')
plt.ylabel(r'$\hat{\mathcal{C}}_L^{\alpha\alpha}\;[\rm{deg}^2]$')
plt.legend()
plt.grid()
```
