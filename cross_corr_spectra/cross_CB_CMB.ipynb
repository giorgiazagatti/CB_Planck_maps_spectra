```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
font = {'family' : 'sans-serif',
            'size'   : 35}
plt.rc('font', **font)
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
```

# $\hat{C}_L^{\alpha X}$ from PR3 data products, with $X = T, E, B$

```python
#file.csv containing multipoles ranging from L=0 up to L=1500 binned each 100 multipoles, the CB x T,E,B spectra presented in Zagatti et al. and the associated error bars
PR3_data_cross = pd.read_csv('PR3_alpha_CMB_cc.csv')

#available spectra and errors
cross_spectra = ['alpha T', 'alpha E', 'alpha B']
errors        = ['err alpha T', 'err alpha E', 'err alpha B']

Lmax = 1500
```

```python
#visualization example of the CBxB power spectrum from Planck PR3
plt.figure(figsize=(15,11))
plt.errorbar(PR3_data_cross['L bin'], PR3_data_cross[cross_spectra[2]], PR3_data_cross[errors[2]], label=cross_spectra[2], fmt='o', lw=2, markersize=15, elinewidth=3, color='black')
plt.hlines(0,1,Lmax,ls='--',color='k')
plt.xlim(0,1500)
plt.ylim(-9e-5, 9e-5)
plt.xlabel('L')
plt.ylabel(r'$\hat{\mathcal{C}}_L^{\alpha B}\;[\rm{deg} \cdot \mu K]$')
plt.legend()
plt.grid()
```
